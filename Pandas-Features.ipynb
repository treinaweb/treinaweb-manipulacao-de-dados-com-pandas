```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Cursos': ["Hadoop", "Spark", "Python", "C#"],
    'Área': ["Data Science", "Data Science", "Programação", "Programação"]
})
df
```

```python
df2 = pd.DataFrame({
    'Cursos': ["Ethereum", "Hyperledger", "Scrum", "XP"],
    'Área': ["Blockchain", "Blockchain", "Engenharia de Software", "Engenharia de Software"]
})
df2
```

# Concatenação Dataframes

```python
data = [df, df2]
df_concat = pd.concat(data)
df_concat
```

```python
df_concat = pd.concat(data, ignore_index=True)
df_concat
```

# Utilizando Joins

```python
tecnologias = {
    'Cursos':[".NET", "Streamlit", "NiFi", "PLN"],
    'Preço':[1000, 800, 950, 750],
    'Duração':['10dias', '6dias', '15dias', '8dias']
}
indices=['l1', 'l2', 'l3', 'l4']
df1 = pd.DataFrame(tecnologias, index=indices)
df1
```

```python
tecnologias2 = {
    'Cursos':[".NET", "Streamlit", "Java", "VC"],
    'Desconto':[200, 50, 250, 150],
}
indices=['l1', 'l6', 'l3', 'l5']
df2 = pd.DataFrame(tecnologias2, index=indices)
df2
```

```python
df3 = df1.join(df2, lsuffix="_left", rsuffix="_right")
df3
```

```python
df3 = df1.join(df2, lsuffix="_left", rsuffix="_right", how="inner")
df3
```

```python
df3 = df1.join(df2, lsuffix="_left", rsuffix="_right", how="left")
df3
```

```python
df3 = df1.join(df2, lsuffix="_left", rsuffix="_right", how="right")
df3
```

```python
df3 = df1.join(df2, lsuffix="_left", rsuffix="_right", how="outer")
df3
```
